```python
final=pd.DataFrame()
for j in range(1,333):
    URL = 'https://www.ambitionbox.com/list-of-companies?campaign=homepage_companies_widget&page={}'.format(j)
    
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
    page = requests.get(URL, headers=headers).text
    
    soup=BeautifulSoup(page,'lxml')
    company = soup.find_all('div', class_='company-content-wrapper')
    
    name = []
    rating=[]
    reviews=[]
    companytype=[]
    headquarters=[]
    old=[]
    employees=[]
    
    for i in company:
        name.append(i.find('h2').text.strip())
        try:
            rating.append(i.find('p', class_='rating').text.strip())
        except:
            rating.append(np.nan)
        try:
            reviews.append(i.find('a', class_='review-count').text.strip())
        except:
            reviews.append(np.nan)
        try:
            companytype.append(i.find_all('p', class_='infoEntity')[0].text.strip())
        except:
            companytype.append(np.nan)
        try:
            headquarters.append(i.find_all('p', class_='infoEntity')[1].text.strip())
        except:
            headquarters.append(np.nan)
        try:
            old.append(i.find_all('p', class_='infoEntity')[2].text.strip())
        except:
            old.append(np.nan)
        try:
            employees.append(i.find_all('p', class_='infoEntity')[3].text.strip())
        except:
            employees.append(np.nan)
    

    d = {'name':name, 'rating':rating, 'reviews':reviews, 'companytype':companytype, 'headquarters':headquarters, 'old':old, 'employees':employees}

    df=pd.DataFrame(d)
    final=final.append(df, ignore_index=True)
```
